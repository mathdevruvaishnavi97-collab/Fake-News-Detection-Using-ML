```python
import pandas as pd
import re
import string
import pickle

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# 2️⃣ Text Cleaning Function
# -----------------------------
def wordopt(text):
    text = str(text).lower()  # lowercase
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # remove URLs
    text = re.sub(r'<.*?>', '', text)  # remove HTML tags
    text = text.translate(str.maketrans('', '', string.punctuation))  # remove punctuation
    text = re.sub(r'\d+', '', text)  # remove numbers
    text = text.strip()
    return text

    
# Apply cleaning to the merged dataset
df_merge["text"] = df_merge["text"].apply(wordopt)
```

```python

# Save trained models
pickle.dump(vectorization, open("vectorization.pkl", "wb"))
pickle.dump(LR, open("LR_model.pkl", "wb"))
pickle.dump(DT, open("DT_model.pkl", "wb"))
pickle.dump(RFC, open("RFC_model.pkl", "wb"))
pickle.dump(GBC, open("GBC_model.pkl", "wb"))
print("Models trained and saved successfully!")
```

```python
# 4️⃣ Load or Train Vectorizer and Models
# -----------------------------
try:
    # Try loading saved vectorizer and models
    with open("vectorization.pkl", "rb") as f:
        vectorization = pickle.load(f)
    with open("LR_model.pkl", "rb") as f:
        LR = pickle.load(f)
    with open("DT_model.pkl", "rb") as f:
        DT = pickle.load(f)
    with open("RFC_model.pkl", "rb") as f:
        RFC = pickle.load(f)
    with open("GBC_model.pkl", "rb") as f:
        GBC = pickle.load(f)
except:
    # If not saved, train small example models (replace with your real training data)
    print("Training models as pickle files not found...")
    x_train = ["fake news example", "real news example"]
    y_train = [0, 1]
    
    # Vectorizer
    vectorization = TfidfVectorizer(max_features=5000)
    x_train_vectors = vectorization.fit_transform(x_train)
    
    # Models
    LR = LogisticRegression()
    LR.fit(x_train_vectors, y_train)
    
    DT = DecisionTreeClassifier()
    DT.fit(x_train_vectors, y_train)
    
    RFC = RandomForestClassifier()
    RFC.fit(x_train_vectors, y_train)
    
    GBC = GradientBoostingClassifier()
    GBC.fit(x_train_vectors, y_train)
    
```

```python
# Save trained models and vectorizer
import pickle
pickle.dump(vectorization, open("vectorization.pkl", "wb"))
pickle.dump(LR, open("LR_model.pkl", "wb"))
pickle.dump(DT, open("DT_model.pkl", "wb"))
pickle.dump(RFC, open("RFC_model.pkl", "wb"))
pickle.dump(GBC, open("GBC_model.pkl", "wb"))

print(" Models and vectorizer saved successfully!")
```

```python
# 1️⃣ Output label function
def output_label(pred):
    if pred == 0:
        return "FAKE"  # 0 = FAKE
    else:
        return "REAL"  # 1 = REAL

# 2️⃣ Ensemble label function
def ensemble_label(preds):
    # Ensure preds are integers: 0 = FAKE, 1 = REAL
    preds = [int(p) for p in preds]
    total = sum(preds)
    
    # Majority vote
    if total > len(preds)/2:
        return "REAL"   # Majority are 1 → REAL
    else:
        return "FAKE"   # Majority are 0 → FAKE

# 3️⃣ Manual testing function
def manual_testing(news):
    testing_news = pd.DataFrame({"text": [news]})
    testing_news["text"] = testing_news["text"].apply(wordopt)
    new_xv_test = vectorization.transform(testing_news["text"])
    
    # Predict using all models
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_RFC = RFC.predict(new_xv_test)
    pred_GBC = GBC.predict(new_xv_test)
    
    print("\nPredictions of individual models:")
    print("Logistic Regression:", output_label(pred_LR[0]))
    print("Decision Tree      :", output_label(pred_DT[0]))
    print("Random Forest      :", output_label(pred_RFC[0]))
    print("Gradient Boosting  :", output_label(pred_GBC[0]))
    
    # Ensemble prediction
    all_preds = [pred_LR[0], pred_DT[0], pred_RFC[0], pred_GBC[0]]
    print("Ensemble Prediction:", ensemble_label(all_preds))
```

```python
news_input = input("Enter news text to test: ")
manual_testing(news_input)
```
